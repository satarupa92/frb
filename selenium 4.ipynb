1.  Scrape the details of most viewed videos on YouTube from Wikipedia: 
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/ 
You need to find following details: 
A) Rank 
B) Name 
C) Artist 
D) Upload date 
E) View

```python
# Getting the chrome-browser 
driver= webdriver.Chrome("chromedriver.exe")
time.sleep(3)

# getting to the link
url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
```

```python
name=[]
try:
    n=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in n:
        name.append(i.text)
except NoSuchElementException:
            name.append('No details available')
except StaleElementReferenceException:
    name.append('No details available')
name    
    
```

```python
Rank=[]
try:
    r=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in r:
        Rank.append(i.text)
except NoSuchElementException:
            Rank.append('No details available')
except StaleElementReferenceException:
    Rank.append('No details available')
Rank    
    
```

```python
artist=[]
try:
    a=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in a:
        artist.append(i.text)
except NoSuchElementException:
            artist.append('No details available')
except StaleElementReferenceException:
    artist.append('No details available')
artist    
```

```python
date=[]
try:
    d=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in d:
        date.append(i.text)
except NoSuchElementException:
            date.append('No details available')
except StaleElementReferenceException:
    date.append('No details available')
date    
```

```python
view=[]
try:
    v=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in v:
        view.append(i.text)
except NoSuchElementException:
            view.append('No details available')
except StaleElementReferenceException:
    view.append('No details available')
view
```

```python
df = pd.DataFrame()
df['rank']=Rank 
df['name']=name
df['artist']=artist
df['views']=view
df['date']=date
df
```

2. Scrape the details team India’s international fixtures from bcci.tv. 
Url = https://www.bcci.tv/. 
You need to find following details: 
A) Match title (I.e. 1st ODI) 
B) Series 
C) Place 
D) Date 
E) Time 
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
# Getting the chrome-browser 
driver= webdriver.Chrome("chromedriver.exe")
time.sleep(3)

# getting to the link
url="https://www.bcci.tv/"
driver.get(url)
```

```python
Int_fixtures=driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']/div/ul/li/a")
try:
    Int_fixtures.click()
except ElementNotInteractableException:
    driver.get(Int_fixtures.get_attribute('href'))
```

```python
match_title=[] 
Series=[]
place=[]
date=[]
time=[]


try:
    m = driver.find_elements_by_xpath("//p[@class='mc-header-info__match-description']/span")
    for i in m:
        match_title.append(i.text)
except NoSuchElementException:
    match_title.append('No details available')

try:
    s=driver.find_elements_by_xpath("//h1[@class='mc-header-info__title']/strong")
    for i in s:
        Series.append(i.text)
except NoSuchElementException:
    Series.append('No details available')
    
    
try:
    p=driver.find_elements_by_xpath("//h1[@class='mc-header-info__title']")
    for i in p:
        place.append(i.text)
except NoSuchElementException:
    place.append('No details available')
    
    
try:
    d=driver.find_elements_by_xpath("//div[@class='mc-header-scorebox__datetime']/strong")
    for i in d:
        date.append(i.text)
except NoSuchElementException:
        date.append('No details available')
        

try:
    t=driver.find_elements_by_xpath("//div[@class='mc-header-scorebox__datetime']/span")
    for i in t:
        time.append(i.text)
except NoSuchElementException:
    time.append('No details available')    
```

```python
df = pd.DataFrame()
df['match_title']=match_title 
df['Series']=Series
df['place']=place
df
```
